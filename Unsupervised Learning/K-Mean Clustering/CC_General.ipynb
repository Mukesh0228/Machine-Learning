```python
import pandas as pd
from sklearn.cluster import KMeans
import  seaborn as sns
from  sklearn.decomposition import PCA
```

```python
df = pd.read_csv("CC_GENERAL.csv")
df
```

```python
df.dropna(axis = 0,inplace=True)
df.drop(columns="CUST_ID",inplace=True)
df.info()
```

```python
model = KMeans(n_clusters=2 , random_state=42)
res = model.fit_predict(df)
```

```python
df["Result"] = res
```

```python
df
```

```python
c1 = df[df["Result"]==1]
c2 = df[df["Result"]==0]
```

```python
model.cluster_centers_
```

```python
res = []
for i in range(1,16):
    model = KMeans(n_clusters=i)
    model.fit_predict(df)
    res.append(model.inertia_)
```

```python
res
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,16),res)
```

```python
sns.scatterplot(df,x="BALANCE",y="PURCHASES",hue ="Result")
```

```python
pca = PCA(n_components=2)
pca_values = pca.fit_transform(df)
pca.explained_variance_
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,RobustScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
main_pipe = Pipeline(
    steps = [
        ("scaling",RobustScaler()),
        ("pca",PCA()),
        ("kmeans",KMeans(n_clusters=2))
    ]
)
```

```python
main_pipe.fit(df)
```

```python
res_pred = main_pipe.fit_predict(df)
```

```python
res_trans = main_pipe.fit_transform(df)
```

```python
sns.scatterplot(res_pred)
```
