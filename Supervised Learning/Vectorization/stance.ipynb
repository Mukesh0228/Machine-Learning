```python
import pandas  as pd
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
from sklearn.metrics import classification_report
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv("mtsamples.csv")
df.dropna(axis = 0,inplace=True)
df
```

```python
df["sentence"] = df["description"] + df["keywords"]
```

```python
x = df.sentence
y = df.medical_specialty
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x, y, random_state = 42, train_size = 0.8)
```

```python
main_pipe  = Pipeline(
    steps =[
        ("vectorizer",CountVectorizer()),
        ("model",LogisticRegression())
    ]
)
```

```python
main_pipe.fit(xtrain,ytrain)
```

```python
ytrain_pred = main_pipe.predict(xtrain)
print(classification_report(ytrain,ytrain_pred))
```

*OBSERVATION*

Logisticregerssion model train perfrom is 99% but test preform only 72% so model comes overfit.

```python
ytest_pred = main_pipe.predict(xtest)
print(classification_report(ytest,ytest_pred))
```

```python
tech = LabelEncoder()
ytrain = tech.fit_transform(ytrain)
ytest = tech.transform(ytest)
```

```python
main_pipe = Pipeline(
    steps=[
        ("vertorize",CountVectorizer()),
        ("model",XGBClassifier())
    ]
)
```

```python
main_pipe.fit(xtrain,ytrain)
```

```python
ytrain_pred = main_pipe.predict(xtrain)
print(classification_report(ytrain,ytrain_pred))
```

```python
ytest_pred = main_pipe.predict(xtest)
print(classification_report(ytest,ytest_pred))
```

```python
main_pipe = Pipeline(
    steps = [
        ("vertorizer",CountVectorizer()),
        ("tree",DecisionTreeClassifier())
    ]
)
```

```python
main_pipe.fit(xtrain,ytrain)
```

```python
ytrain_pred = main_pipe.predict (xtrain)
print(classification_report(ytrain,ytrain_pred))
```

```python
ytest_pred = main_pipe.predict(xtest)
print(classification_report(ytest,ytest_pred))
```
